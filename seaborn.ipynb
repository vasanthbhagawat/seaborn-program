```python
tips
```

```python
sns.scatterplot(x="tip",y='total_bill',data=tips)
```

```python
sns.scatterplot(x="tip",y='total_bill',data=tips,hue='day')
```

```python
sns.scatterplot(x="tip",y='total_bill',data=tips,hue='day',size="sex")
```

# 2. histogram

```python
sns.histplot(tips['tip'],kde=True)
```

```python
sns.histplot(tips['tip'],kde=True,bins=25)
```

```python
sns.histplot(tips['tip'],kde=True,bins=25).set(title="histogram plot")
```

# bar plot

```python
sns.barplot(x="time",y="tip",data=tips).set(title="bar plot")
```

```python
sns.boxplot(x="day",y="tip",data=tips)
```

# Box plot

```python
sns.boxplot(x="day",y="tip",data=tips,hue="sex")
```

# Strip Plot

```python
sns.stripplot(x='day',y='tip',data=tips)
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue="sex")
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue="sex",dodge=True)
```

# Jointplot

```python
sns.jointplot(x="tip",y="total_bill",data=tips)
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="reg")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde",shade=True)
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="hex",color='yellow')
```
